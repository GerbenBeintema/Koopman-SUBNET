```python
# import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import deepSI
# import ipywidgets as widgets
from scipy.io import loadmat # to load .mat data

# %matplotlib widget
```

```python
# load data
# load external data
Tot = deepSI.datasets.Silverbox(split_data=False)[1]
uTot = Tot.u[None]
yTot = Tot.y[None]

NTrain = 69500
NVal = 8500
NTest = NVal
iTrainStart = 40551
iValStart = 110070
iTestStart = 118775

uTrain = uTot[0][iTrainStart:iTrainStart+NTrain]
uVal = uTot[0][iValStart:iValStart+NVal]
uTest = uTot[0][iTestStart:iTestStart+NTest]
uTestArrow = uTot[0][0:iTrainStart]

yTrain = yTot[0][iTrainStart:iTrainStart+NTrain]
yVal = yTot[0][iValStart:iValStart+NVal]
yTest = yTot[0][iTestStart:iTestStart+NTest]
yTestArrow = yTot[0][0:iTrainStart]
```

```python
# some variables
nf = 200 # length of truncated simulation
nx = 20 # number of states
nb = 5 # input history for encoder
na = 5 # output history for encoder
nne = 40 # number of neurons per layer for encoder
nng = 40 # number of neurons per layer for affine input matrix function
nnh = 20 # number of neurons per layer for output function
nle = 1 # number of layers encoder
nlg = 1 # number of layers state transition function
nlh = 1 # number of layers output function

trainEpochs = 10000
```

```python
# create training validation test data structures

# create deepSI data structure
trainData = deepSI.System_data(u=uTrain,y=yTrain)
valData = deepSI.System_data(u=uVal,y=yVal)
testData = deepSI.System_data(u=uTest,y=yTest)
testArrowData = deepSI.System_data(u=uTestArrow,y=yTestArrow)
```

```python
# plot data
plt.figure()
plt.plot(trainData.u)
plt.ylabel('u'); plt.xlabel('time'); plt.legend(['train input'])
plt.show()

# plt.figure()
# plt.plot(trainData.y)
# plt.ylabel('y'); plt.xlabel('time'); plt.legend(['train output'])
# plt.show()
                                        
plt.figure()
plt.plot(valData.u)
plt.ylabel('u'); plt.xlabel('time'); plt.legend(['validation input'])
plt.show()

# plt.figure()
# plt.plot(valData.y)
# plt.ylabel('y'); plt.xlabel('time'); plt.legend(['validation output'])
# plt.show()

plt.figure()
plt.plot(testArrowData.u)
plt.ylabel('u'); plt.xlabel('time'); plt.legend(['test input'])
plt.show()

# plt.figure()
# plt.plot(testData.y)
# plt.ylabel('y'); plt.xlabel('time'); plt.legend(['test output'])
# plt.show()
```

```python
# init model structure
mod_KoopmanEncoder = deepSI.fit_systems.SS_encoder_affine_input(nx=nx, na=na, nb=nb, 
                                                    e_net_kwargs={'n_nodes_per_layer':nne, 'n_hidden_layers':nle}, 
                                                    g_net_kwargs={'n_nodes_per_layer':nng, 'n_hidden_layers':nlg}, 
                                                    h_net_kwargs={'n_nodes_per_layer':nnh, 'n_hidden_layers':nlh})
```

```python
# estimate model using SS-Encoder method

mod_KoopmanEncoder.fit(trainData, epochs=trainEpochs, batch_size=256, loss_kwargs={'nf':nf,'dilation':1}, sim_val=valData, concurrent_val=True)
# sys_encoder.fit(trainData, epochs=trainEpochs, batch_size=256, loss_kwargs={'nf':nf,'dilation':1})

# plot validation error evolution
plt.figure(facecolor='w', edgecolor='k')
plt.semilogy(mod_KoopmanEncoder.Loss_val[:-1])
plt.semilogy(mod_KoopmanEncoder.Loss_val[-trainEpochs:-1])
plt.show()
```

```python
# mod_KoopmanEncoder.save_system('KoopmanAffine_Silverbox_v1.sav')
mod_KoopmanEncoder = deepSI.load_system('KoopmanAffine_Silverbox_v1.sav')
```

```python
# validate &plot results
# plot validation error evolution
# validate & plot results
testArrow_simulation_encoder = mod_KoopmanEncoder.apply_experiment(testArrowData)
testArrow075_simulation_encoder = mod_KoopmanEncoder.apply_experiment(testArrowData[0:30000])
test_simulation_encoder = mod_KoopmanEncoder.apply_experiment(testData)
val_simulation_encoder = mod_KoopmanEncoder.apply_experiment(valData)
train_simulation_encoder = mod_KoopmanEncoder.apply_experiment(trainData)
```

```python
# display
print('  Encoder-Koopman Affine  Model')
print('NRMS Error')
print('Multisine Train: ' + str(train_simulation_encoder.NRMS(trainData)))
print('Multisine Val: ' + str(val_simulation_encoder.NRMS(valData)))
print('Multisine Test: ' + str(test_simulation_encoder.NRMS(testData))) 
print('Arrow Test: ' + str(testArrow_simulation_encoder.NRMS(testArrowData))) 
print('Arrow Test - no extrapol: ' + str(testArrow075_simulation_encoder.NRMS(testArrowData[0:30000]))) 
print(' ')
print('RMS Error')
print('Multisine Train: ' + str(train_simulation_encoder.RMS(trainData)*1000))
print('Multisine Val: ' + str(val_simulation_encoder.RMS(valData)*1000))
print('Multisine Test: ' + str(test_simulation_encoder.RMS(testData)*1000)) 
print('Arrow Test: ' + str(testArrow_simulation_encoder.RMS(testArrowData)*1000)) 
print('Arrow Test - no extrapol: ' + str(testArrow075_simulation_encoder.RMS(testArrowData[0:30000])*1000)) 

scale = 0.6
Ts = 1/610.35
plot = lambda x: plt.plot(np.arange(len(x))*Ts, x)

plt.figure(facecolor='w', edgecolor='k', figsize=(15*scale,4*scale))
plt.subplot(1,2,1)
plot(testArrowData.y)
plot(testArrowData.y-testArrow_simulation_encoder.y)
plt.ylabel('y (V)'); plt.xlabel('time (sec)'); #plt.legend(['test output', 'simulation error test encoder'])
plt.grid()


plt.subplot(1,2,2)
# plt.figure(facecolor='w', edgecolor='k')
plot(testData.y)
plot(testData.y-test_simulation_encoder.y)
# plt.ylabel('y'); 
plt.xlabel('time (sec)'); plt.legend(['Test Measurements', 'Koopman SUBNET simulation error']); plt.grid()
plt.tight_layout(pad=0.2)
plt.savefig('../../../figures/koopman-silverbox-results.pdf')
plt.show()
```
